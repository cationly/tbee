## dimer chain

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}, 
             {'tag': b'b', 'r0': [1, 0]}]
prim_vec = {'norm': 2, 'angle': 0}

lat = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt = plot(lat=lat)
sav = save(lat=lat, dir_name='chain')

n1, n2 = 10, 1
lat.get_lattice(n1=n1, n2=n2)
fig_lat = plt.lattice(ms=20, figsize=(10, 2))
sav.fig_lat(fig_lat, 'lattice')
```

## Square lattice 

###  with 1 site within the unit cell

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}]
prim_vec = {'norm': 2, 'angle': 90}

lat = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt = plot(lat=lat)
sav = save(lat=lat, dir_name='square')

n1, n2 = 8, 8
lat.get_lattice(n1=n1, n2=n2)
fig_lat = plt.lattice(ms=15)
sav.fig_lat(fig_lat, 'lattice')
```

## Face-centered square

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}, 
             {'tag': b'b', 'r0': [1, 1]}]
prim_vec = {'norm': 2, 'angle': 90}

lat = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt = plot(lat=lat)
sav = save(lat=lat, dir_name='face_center_square')

n1, n2 = 8, 8
lat.get_lattice(n1=n1, n2=n2)
lat.boundary_line(cx=0, cy=-1, co=-12.5)
lat.boundary_line(cx=-1, cy=0, co=-12.5)
fig_lat = plt.lattice(ms=15)
sav.fig_lat(fig_lat, 'lattice')
```

## Lieb lattice

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]},
             {'tag': b'b', 'r0': [1, 0]},
             {'tag': b'c', 'r0': [0, 1]}]
prim_vec = {'norm': 2, 'angle': 90}
lat = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt = plot(lat=lat)
sav = save(lat=lat, dir_name='lieb')

n1, n2 = 6, 6
lat.get_lattice(n1=n1, n2=n2)
lat.remove_dangling()
fig_lat = plt.lattice(ms=15)
sav.fig_lat(fig_lat, 'lattice')
```

## graphene

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}, 
             {'tag': b'b', 'r0': [0.5*sqrt(3), 0.5]}]
prim_vec = {'norm': sqrt(3), 'angle': 60}

lat = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt = plot(lat=lat)
sav = save(lat=lat, dir_name='graphene')

n1, n2 = 6, 6
lat.get_lattice(n1=n1, n2=n2)
fig_lat_ind = plt.lattice(ms=15, plt_index=True)
sav.fig_lat(fig_lat_ind, 'lattice_ind')
```

```python
lat.remove_sites(index=[27, 32, 33, 38, 39, 44])
lat.remove_dangling()
fig_lat_vac = plt.lattice(ms=15)
sav.fig_lat(fig_lat_vac, 'lattice_index_vac')
```

## Kagome lattice

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}, 
             {'tag': b'b', 'r0': [1, 0]},
             {'tag': b'c', 'r0': [1/2., sqrt(3)/2]}]
prim_vec = {'norm': 2, 'angle': 60}

lat = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt = plot(lat=lat)
sav = save(lat=lat, dir_name='kagome')

n1, n2 = 6, 5
lat.get_lattice(n1=n1, n2=n2)
fig_lat = plt.lattice(ms=15)
sav.fig_lat(fig_lat, 'lattice')
```

## Line centered hexagon 

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}, 
             {'tag': b'a', 'r0': [0.5*sqrt(3), 0.5]}, 
             {'tag': b'b', 'r0': [0.25*sqrt(3), 0.25]},
             {'tag': b'b', 'r0': [0.75*sqrt(3), 0.25]},
             {'tag': b'b', 'r0': [0.5*sqrt(3), 1.]}]
prim_vec = {'norm': sqrt(3), 'angle': 60}
lat = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt = plot(lat=lat)
sav = save(lat=lat, dir_name='graphene_line_center')

n1, n2 = 6, 6
lat.get_lattice(n1=n1, n2=n2)
lat.remove_dangling()
fig_lat_ind = plt.lattice(ms=10)
sav.fig_lat(fig_lat_ind, 'lattice')
```

## ELLIPSOID SHAPES

### Circle

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}]
prim_vec = {'norm': 1, 'angle': 90}
lat = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt = plot(lat=lat)
n1, n2 = 12, 12
lat.get_lattice(n1=n1, n2=n2)
lat.center()
lat.ellipse_in(a=5.8, b=5.8)
fig_lat = plt.lattice(ms=10)
```

### Ellipse

### circle removed

### Circular graphene flake 

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}, 
             {'tag': b'a', 'r0': [0.5*sqrt(3), 0.5]}]
prim_vec = {'norm': sqrt(3), 'angle': 60}
lat = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt = plot(lat=lat)
sav = save(lat=lat, dir_name='hexa_line_center')

n1, n2 = 16, 16
lat.get_lattice(n1=n1, n2=n2)
lat.center()
lat.ellipse_in(a=10, b=10)
lat.remove_dangling()
fig_lat = plt.lattice(ms=5, tics=True)
```

## REMOVE SITES

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}]
prim_vec = {'norm': 1, 'angle': 60}
lat1 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt1 = plot(lat=lat1)
n1, n2 = 7, 7
lat1.get_lattice(n1=n1, n2=n2)
fig_lat1 = plt1.lattice(ms=20, tics=True)

lat2 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt2 = plot(lat=lat2)
n1, n2 = 3, 3
lat2.get_lattice(n1=n1, n2=n2)
lat2.shift_x(shift=3)
lat2.shift_y(shift=sqrt(3))
fig_lat2 = plt2.lattice(ms=20, tics=True)

lat1 -= lat2
plt1.lattice(ms=20, tics=True)
plt.show()
```

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}, 
                 {'tag': b'b', 'r0': [0.5*sqrt(3), 0.5]}]
prim_vec = {'norm': sqrt(3), 'angle': 60}
lat0 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt0 = plot(lat=lat0)
n1, n2 = 5, 8
lat0.get_lattice(n1=n1, n2=n2)
plt0.lattice(tics=True);

lat1 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt1 = plot(lat=lat1)
n1, n2 = 2, 2
lat1.get_lattice(n1=n1, n2=n2)
plt1.lattice(tics=True);

lat2 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt2 = plot(lat=lat2)
lat2.get_lattice(n1=n1, n2=n2)
lat2.shift_x(2*sqrt(3))
lat2.shift_y(3)
plt2.lattice(tics=True);

lat3 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt3 = plot(lat=lat3)
lat3.get_lattice(n1=n1, n2=n2)
lat3.shift_x(4*sqrt(3))
lat3.shift_y(6)
plt3.lattice(tics=True);

lat = lat0 - lat1 - lat2 - lat3
plt = plot(lat=lat)
plt.lattice();
```

## ADD SITES

```python
# run previous cell first
lat = lat3 + lat2 + lat1
plt = plot(lat=lat)
plt.lattice();
```

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}]
prim_vec = {'norm': 1, 'angle': 90}

# lattice 1 (square)
lat1 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt1 = plot(lat=lat1)
sav1 = save(lat=lat1, dir_name='square')
n1, n2 = 7, 7
lat1.get_lattice(n1=n1, n2=n2)
fig_lat = plt1.lattice(ms=15, tics=True)

# lattice 2 (rectangle)
lat2 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt2 = plot(lat=lat2)
n1, n2 = 6, 3
lat2.get_lattice(n1=n1, n2=n2)
lat2.shift_x(shift=6)
lat2.shift_y(shift=2)
fig_lat = plt2.lattice(ms=15, tics=True)

lat1 += lat2
plt1.lattice(ms=15, tics=True);

# lattice 3 (square)
lat3 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt3 = plot(lat=lat3)
sav3 = save(lat=lat3, dir_name='square')
n1, n2 = 7, 7
lat3.get_lattice(n1=n1, n2=n2)
lat3.shift_x(shift=12)

lat1 += lat3
plt1.lattice(ms=15, tics=True);
```

```python
unit_cell = [{'tag': b'a', 'r0': [0, 0]}, 
                 {'tag': b'b', 'r0': [0.5*sqrt(3), 0.5]}]
prim_vec = {'norm': sqrt(3), 'angle': 60}
lat1 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt1 = plot(lat=lat1)

# rhombus = 2 triangles
n = 10 # no of plackets along one edge.
lat1.get_lattice(n1=n+2, n2=n+2)
lat1.boundary_line(cx=-sqrt(3), cy=-1, co=-3*(n+1))
fig_lat = plt1.lattice(ms=10, tics=True)

unit_cell = [{'tag': b'b', 'r0': [0, 0]}, 
                 {'tag': b'a', 'r0': [0.5*sqrt(3), 0.5]}]
prim_vec = {'norm': sqrt(3), 'angle': 60}

lat2 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt2 = plot(lat=lat2)
lat2.get_lattice(n1=n+2, n2=n+2)
lat2.boundary_line(cx=-sqrt(3), cy=-1, co=-3*(n+1))
lat2.factor_y(factor=-1)
lat2.shift_y(shift=-1)
fig_lat = plt2.lattice(ms=10, tics=True)

lat1 += lat2 
fig_lat = plt1.lattice(ms=10, figsize=(16, 10))
```

# LOGO

```python
n1, n2 = 5, 4
unit_cell = [{'tag': b'a', 'r0': [0, 0]}]
prim_vec = {'norm': 1, 'angle': 0}
lat1 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt1 = plot(lat=lat1)

lat1.get_lattice(n1=n1, n2=1)
plt1.lattice(figsize=(4, 3), tics=True);

unit_cell = [{'tag': b'a', 'r0': [2, -4]}]
prim_vec = {'norm': 1, 'angle': 90}
lat2 = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
plt2 = plot(lat=lat2)

lat2.get_lattice(n1=1, n2=n2)
plt2.lattice(figsize=(4, 3), tics=True);

lat1 += lat2
lat1.shift_y(4)
lat1.shift_x(-2)
plt1.lattice(tics=True);
```
