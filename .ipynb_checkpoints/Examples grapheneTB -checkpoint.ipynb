# GRAPHENE FLAKES

## Quick example

```python
from grapheneTB import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%load_ext autoreload
%autoreload 2
```

```python
lat = grapheneTB()
lat.get_lattice(nx=6, ny=6)
lat.plt_lattice(ms=4);
```

```python
sys = eigTB(lat)
sys.print_hop(n=2)
```

```python
sys.set_hop_uni([{'n':1, 't':1.}])
```

```python
sys.rename_hop_tag([{'n':1, 'ang':90, 'tag_new':'u'}, 
                    {'n':1,'ang':30, 'tag_new':'v'}, 
                    {'n':1,'ang':150, 'tag_new':'w'}])
```

```python
sys.set_hop_with_tag([{'tag':b'u', 't':1}, {'tag':b'v', 't':2}, {'tag':b'w', 't':3}])
```

```python
sys.get_coor_hop()
```

```python
pl = plotTB(sys)
pl.plt_lattice_hop(ms=5);
```

```python
sys.get_ham(complex_transpose=True)
sys.get_eig(eigenvec=True)
```

```python
pl.plt_spec(tag_pola=b'a');
```

# GRAPHENE FLAKES UNDER LINEAR TRIAXIAL STRAIN

## quick example

```python
lat = grapheneTB()
lat.triangle_zigzag(14)
lat.plt_lattice(ms=4);
```

```python
sys = grapheneEig(lat)
pl = plotTB(sys)
sys.set_hop_uni([{'n':1, 't':1.}])
```

```python
sys.get_ham(complex_transpose=True)
sys.get_eig(eigenvec=True)
pl.plt_spec(tag_pola=b'a');
```

```python
zero_modes = sys.get_states_en(en_lims=[-0.1, 0.1])
pl.plt_intensity_disk(zero_modes, s=2000, plt_hop=True);
```

## Strained Hexagonal Flakes

```python
lat = grapheneTB()
lat.hexagon_zigzag(5)
sys = grapheneEig(lat)
pl = plotTB(sys)
betas = sys.get_beta_lims()
```

### Increasing negative strain

```python
sys.set_hop_linear_strain(t=1, beta=0)
sys.get_coor_hop()
pl.plt_lattice_hop(ms=5, lw=2, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=1/4*betas[0])
sys.get_coor_hop()
pl.plt_lattice_hop(ms=5, lw=2, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=2/4*betas[0]);
sys.get_coor_hop()
pl.plt_lattice_hop(ms=5, lw=2, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=3/4*betas[0])
sys.get_coor_hop()
pl.plt_lattice_hop(ms=5, lw=2, plt_hop=True, figsize=(3, 3));
```

### Increasing positive strain

```python
sys.set_hop_linear_strain(t=1, beta=0)
sys.get_coor_hop()
pl.plt_lattice_hop(ms=5, lw=2, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=1/4*betas[1])
sys.get_coor_hop()
pl.plt_lattice_hop(ms=5, lw=2, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=2/4*betas[1])
sys.get_coor_hop()
pl.plt_lattice_hop(ms=5, lw=2, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=3/4*betas[1])
sys.get_coor_hop()
pl.plt_lattice_hop(ms=5, lw=2, plt_hop=True, figsize=(3, 3));
```

## Pseudo Landau levels in the triangular flake with zigzag terminations 

```python
lat = grapheneTB()
lat.triangle_zigzag(24)
sys = grapheneEig(lat)
plt = plotTB(sys)
betas = sys.get_beta_lims()
sys.set_hop_linear_strain(t=1, beta=betas[1])
sys.get_coor_hop()
plt.plt_lattice(ms=5);
plt.plt_lattice_hop(ms=5);
```

```python
sys.get_ham(complex_transpose=True)
sys.get_eig(eigenvec=True)
plt.plt_spec(tag_pola=b'a', en_lims=[-2.2, 2.2]);
```

```python
ll0 = sys.get_states_en(en_lims=[-0.1, 0.1])
ll1 = sys.get_states_en(en_lims=[0.8, 1.2])
ll2 = sys.get_states_en(en_lims=[1.2, 1.4])
ll3 = sys.get_states_en(en_lims=[1.4, 1.6])
ll4 = sys.get_states_en(en_lims=[1.6, 1.8])
ll5 = sys.get_states_en(en_lims=[1.8, 2.])
plt.plt_intensity_disk(ll0, s=2000, plt_hop=True);
plt.plt_intensity_disk(ll1, s=2000, plt_hop=True);
plt.plt_intensity_disk(ll2, s=2000, plt_hop=True);
```

```python
plt.plt_intensity_disk(ll3, s=2000, plt_hop=True);
plt.plt_intensity_disk(ll4, s=2000, plt_hop=True);
plt.plt_intensity_disk(ll5, s=2000, plt_hop=True);
```

## Vacancy defect leading to localized states between Laudau levels 

### site A at the center of mass removed

```python
lat = grapheneTB()
lat.triangle_zigzag(25)
lat.remove_sites([lat.sites//2])
sys = grapheneEig(lat)
plt = plotTB(sys)
betas = sys.get_beta_lims()
sys.set_hop_linear_strain(t=1, beta=betas[1])
plt.plt_lattice(ms=5);
```

```python
sys.get_ham(complex_transpose=True)
sys.get_eig(eigenvec=True)
plt.plt_spec(ms=8, tag_pola=b'a', en_lims=[-2.2, 2.2]);
```

```python
ll0 = sys.get_states_en(en_lims=[-0.2, 0.2])
ll1 = sys.get_states_en(en_lims=[.8, 1.])
ll2 = sys.get_states_en(en_lims=[1.2, 1.3])
v0 = sys.get_states_en(en_lims=[0.2, 0.6])
v1 = sys.get_states_en(en_lims=[1., 1.05])
v2 = sys.get_states_en(en_lims=[1.35, 1.4])
```

```python
plt.plt_intensity_disk(ll0, s=2000, plt_hop=True);
plt.plt_intensity_disk(ll1, s=2000, plt_hop=True);
plt.plt_intensity_disk(ll2, s=2000, plt_hop=True);
```

```python
plt.plt_intensity_disk(v0, s=5000, plt_hop=True);
plt.plt_intensity_disk(v1, s=5000, plt_hop=True);
plt.plt_intensity_disk(v2, s=5000, plt_hop=True);
```

## Vacancy defect leading to localized states between Laudau levels

```python
lat = grapheneTB()
lat.triangle_zigzag(24)
sys = grapheneEig(lat)
plt = plotTB(sys)
betas = sys.get_beta_lims()
sys.set_hop_linear_strain(t=1, beta=betas[1])
sys.set_vortex()
sys.get_ham(complex_transpose=True)
sys.get_eig(eigenvec=True)
plt.plt_spec(tag_pola=b'a', en_lims=[-2.2, 2.2]);
```

```python
ll0 = sys.get_states_en(en_lims=[-0.2, 0.2])
ll1 = sys.get_states_en(en_lims=[.8, 1.])
ll2 = sys.get_states_en(en_lims=[1.2, 1.4])
v0 = sys.get_states_en(en_lims=[.4, .8])
v1 = sys.get_states_en(en_lims=[1., 1.2])
v2 = sys.get_states_en(en_lims=[1.4, 1.51])
plt.plt_intensity_disk(ll0, s=5000, plt_hop=True);
plt.plt_intensity_disk(ll1, s=5000, plt_hop=True);
plt.plt_intensity_disk(ll2, s=5000, plt_hop=True);
```

```python
plt.plt_intensity_disk(v0, s=10000, plt_hop=True);
plt.plt_intensity_disk(v1, s=10000, plt_hop=True);
plt.plt_intensity_disk(v2, s=10000, plt_hop=True);
```

## Hoftstater Butterfly

### triangular flake

```python
lat = grapheneTB()
lat.triangle_zigzag(24)
sys = grapheneEig(lat)
sys.get_butterfly(t=1, N=30)
plt = plotTB(sys)
plt.plt_butterfly();
```

### hexagonal flake

```python
lat = grapheneTB()
lat.hexagon_zigzag(10)
sys = grapheneEig(lat)
sys.get_butterfly(t=1, N=30)
plt = plotTB(sys)
plt.plt_butterfly();
```

### graphene flake

```python
lat = grapheneTB()
lat.circle(20)
sys = grapheneEig(lat)
sys.get_butterfly(t=1, N=30)
plt = plotTB(sys)
plt.plt_butterfly()
plt.show()
```
