# PT-symmetry / PT-symmetry breaking

These examples illustrate the PT-symmetry and the PT-symmetry breaking 
considering a dimer and a dimer chain. 

```python
%load_ext autoreload
%autoreload 2
```

```python
from chain_tb import *
%matplotlib inline
```

### Lattice parameters

```python
tags = [b'a', b'b']
ri = [[0, 0], [1, 0]]
pv = [[2, 0], [2, 0]]
figsize = (8, 3)
ta = 1.
```

## Dimer

Throughout these examples, the hopping $t$ between the A and B site is real and the onsite energies are complex conjugated and purely imaginary with $\epsilon_a=i\gamma$ and $\epsilon_b=-i\gamma$. Note that the A site is amplified (positive imaginary part) and the B site is absorbed.

The Hamiltian of the dimer reads:



$$\left( \begin{array}{cc}
  i\gamma & t \\
  t & i\gamma \end{array} \right) $$

```python
n = 2
```

```python
chain_tb = latticeTB(tags=tags, ri=ri, pv=pv)
chain_tb.get_lattice(nx=n, ny=1)
```

```python
eig_chain = eigChain(chain_tb.coor, chain_tb.tags)
eig_chain.set_hop(ta=ta)
```

```python
plot_chain = plotChain(chain_tb.coor, chain_tb.tags)
fig_lattice = plot_chain.plt_lattice(ms=60, hop=eig_chain.hop, c=10, label=False)
```

### Hermitian Case

In the case $\gamma=0$, the eigenenergies are real and given by: $E_\pm=\pm t$.

```python
eig_chain.get_ham()
eig_chain.get_eig()
```

```python
fig_spec = plot_chain.plt_spec(eig_chain.en, eig_chain.pola,
                               ms=20, pola_tag=b'a')
```

```python
# Time Evolution
prop_chain = propagationTB(coor=chain_tb.coor, steps=1400, dz=0.01)
psi_init = np.zeros(prop_chain.sites, 'c16')
psi_init[0] = 1.
prop_chain.get_prop(ham=eig_chain.ham, psi_init=psi_init, norm=True)
fig_ints = prop_chain.plt_prop_dimer()
```

```python
fig_prop = prop_chain.plt_prop1d()
```

## Non-Hermitian case

In the case $\gamma\neq 0$, the eigenenergies are given by: 
$E_\pm=\pm \sqrt{t^2-|\gamma|^2}$. They are real if $t>\gamma$, degenerated if $t=\gamma$, and complex if $t<\gamma$.

Note that a non-Hermitian matrix can then have a real spectrum.

If the eigenenergies are real ($t>\gamma$) the system displays the PT-symmetry, if they are complex, the PT-symmetry is broken ($t<\gamma$), $\gamma=t$ represents the PT-symmetry breaking point. In the following, we will consider each case separately.

```python
onsite = [0.8j, -0.8j] 
eig_chain.set_onsite(onsite)
eig_chain.get_ham()
eig_chain.get_eig()
```

### PT-symmetry
Below the PT-symmetry breaking point, the eigenenergies are real: and given by: $E_\pm=\pm \sqrt{t^2-|\gamma|^2}$.

```python
fig_spec = plot_chain.plt_spec(eig_chain.en, eig_chain.pola,
                               ms=20, pola_tag=b'a')
```

```python
# Time Evolution
prop_chain = propagationTB(coor=chain_tb.coor, steps=1400, dz=0.01)
psi_init = np.zeros(prop_chain.sites, 'c16')
psi_init[0] = 1.
prop_chain.get_prop(ham=eig_chain.ham, psi_init=psi_init, norm=True)
fig_ints = prop_chain.plt_prop_dimer()
```

```python
fig_prop = prop_chain.plt_prop1d()
```

### PT-symmetry breaking point

If $\gamma=t$, the eigenenergies coincide: $E_\pm=0$.

```python
onsite = [1j, -1j] 
eig_chain.set_onsite(onsite)
eig_chain.get_ham()
eig_chain.get_eig()
```

```python
fig_spec = plot_chain.plt_spec(eig_chain.en, eig_chain.pola,
                               ms=20, pola_tag=b'a')
```

```python
# Time Evolution
prop_chain = propagationTB(coor=chain_tb.coor, steps=1400, dz=0.01)
psi_init = np.zeros(prop_chain.sites, 'c16')
psi_init[0] = 1.
prop_chain.get_prop(ham=eig_chain.ham, psi_init=psi_init, norm=True)
fig_ints = prop_chain.plt_prop_dimer()
```

```python
fig_prop = prop_chain.plt_prop1d()
```

### Broken PT-symmetry
Above the PT-symmetry breaking point, $\gamma>t$, the eigenenergies are purely imaginary: $E_\pm= \pm i \sqrt{\gamma^2-t^2}$.

```python
onsite = [1.2j, -1.2j] 
eig_chain.set_onsite(onsite)
eig_chain.get_ham()
eig_chain.get_eig()
```

```python
fig_spec = plot_chain.plt_spec(eig_chain.en, eig_chain.pola,
                               ms=20, pola_tag=b'a')
```

Time evolution with initial condition $\psi_i = (1, 0)$

```python
prop_chain = propagationTB(coor=chain_tb.coor, steps=1400, dz=0.01)
psi_init = np.zeros(prop_chain.sites, 'c16')
psi_init[0] = 1.
prop_chain.get_prop(ham=eig_chain.ham, psi_init=psi_init, norm=True)
fig_ints = prop_chain.plt_prop_dimer()
```

```python
fig_prop = prop_chain.plt_prop1d()
```

Time evolution with initial condition $\psi_i = (0, 1)$

```python
psi_init = np.zeros(prop_chain.sites, 'c16')
psi_init[1] = 1.
prop_chain.get_prop(ham=eig_chain.ham, psi_init=psi_init, norm=True)
fig_ints = prop_chain.plt_prop_dimer()
```

```python
fig_prop = prop_chain.plt_prop1d()
```

Note that the final state does not depend of the initial conditions.

## Dimer Chain

The chain is composed of $n$ dimers i.e. $n$ sites A and $n$ sites B. The chain starts with a A site and ends with a B site.

```python
n = 20
```

```python
chain_tb.get_lattice(nx=n, ny=1)
eig_chain = eigChain(chain_tb.coor, chain_tb.tags)
eig_chain.set_hop(ta)
```

```python
plot_chain = plotChain(chain_tb.coor, chain_tb.tags)
fig_lattice = plot_chain.plt_lattice(ms=12, c=4,hop=eig_chain.hop,
                                     numero=False, figsize=figsize)
```

### Hermitian Case

```python
eig_chain.get_ham()
eig_chain.get_eig()
```

```python
fig_spec = plot_chain.plt_spec(eig_chain.en, eig_chain.pola,
                               ms=20, pola_tag=b'a')
```

The chain can be viewed as a conductor in the condensed-matter language (no gap between the negative and positive energies). The spectrum displays the particle-hole symmetry: $\{\pm |E_n|\}$.

There is no states of zero energy and the polarization shows that all the states of the chain are equaly distributed on each sublattice, $\langle A|A \rangle=\langle B|B \rangle=1/2$.

```python
prop_chain = propagationTB(coor=chain_tb.coor, steps=1000, dz=0.1)
psi_init = np.ones(prop_chain.sites, 'c16')/np.sqrt(prop_chain.sites)
prop_chain.get_prop(ham=eig_chain.ham, psi_init=psi_init, norm=True)
fig_prop = prop_chain.plt_prop_chain()
```

### Below the PT-symmetry breaking 

```python
onsite = [0.1j, -0.1j]
eig_chain.set_onsite(onsite)
eig_chain.get_ham()
eig_chain.get_eig()
```

```python
fig_spec = plot_chain.plt_spec(eig_chain.en, eig_chain.pola,
                               ms=20, pola_tag=b'a')
```

```python
prop_chain = propagationTB(coor=chain_tb.coor, steps=1000, dz=0.1)
psi_init = np.ones(prop_chain.sites, 'c16')/np.sqrt(prop_chain.sites)
prop_chain.get_prop(ham=eig_chain.ham, psi_init=psi_init, norm=True)
fig_prop = prop_chain.plt_prop_chain()
```

### Above the PT-symmetry breaking 

```python
onsite = [0.2j, -0.2j]
eig_chain.set_onsite(onsite)
eig_chain.get_ham()
eig_chain.get_eig()
```

```python
fig_spec = plot_chain.plt_spec(eig_chain.en, eig_chain.pola,
                               ms=20, pola_tag=b'a')
```

 A pair of complex-conjugate eigenenergies appears. Note that these states do not statisfy $\langle A|A \rangle \ne \langle B|B \rangle$. One state is then be amplified (the one this $\langle A|A \rangle > 1/2$) and the other one absorbed.

```python
state_ampl = eig_chain.get_state_pola('a')
fig_ampl_state = plot_chain.plt_intensity(state_ampl, add='ampl state')
```

```python
prop_chain = propagationTB(coor=chain_tb.coor, steps=1000, dz=0.1)
psi_init = np.ones(prop_chain.sites, 'c16')/np.sqrt(prop_chain.sites)
prop_chain.get_prop(ham=eig_chain.ham, psi_init=psi_init, norm=True)
fig_prop = prop_chain.plt_prop_chain()
```
