```python
lat = grapheneLat()
sys = system(lat=lat)
plt = plot(sys=sys)
lat.get_lattice(n1=6, n2=6)
lat.remove_dangling()
lat.center()

lat.remove_sites([26])


sites_no_impurity = lat.sites
coor = np.array([(0, -1, b'c'), (0, -1, b'd')], dtype=[('x', 'f8'), ('y', 'f8'), ('tag', 'S1')])
lat.add_sites(coor=coor)

lat.plot(axis=True, plt_index=True, ms=5, fs=10);
```

```python
sys.print_distances(3)
```

```python
sys.set_onsite({b'a': 1, b'b': 1})
```

```python
# Energies of the 2 levels
sys.set_onsite({b'c': 1, b'd': 2})
```

```python
sys.set_hopping([{'n': 1, 't': 1, 'tag': b'ab'},
                 {'n': 1, 't': 1, 'tag': b'ba'}])

sys.set_hopping([{'n': 2, 't': 0.8, 'tag': b'aa'},
                 {'n': 2, 't': 0.8, 'tag': b'bb'}])
```

```python
#Hoppings to/from impurity
# NEAREST NEIGHBOURS HOPPINGS
t0 = 0.6
t1 = 0.4
sys.set_hopping([{'n': 1, 't': t0, 'tag': b'cb'}, 
                 {'n': 1, 't': t0, 'tag': b'bc'}])
sys.set_hopping([{'n': 1, 't': t1, 'tag': b'db'},
                 {'n': 1, 't': t1, 'tag': b'bd'}])
# NEXT NEAREST NEIGHBOURS HOPPINGS
t2 = 0.4
t3 = 0.3
sys.set_hopping([{'n': 2, 't': t2, 'tag': b'ca'}, 
                 {'n': 2, 't': t2, 'tag': b'ac'}])
sys.set_hopping([{'n': 2, 't': t3, 'tag': b'da'},
                 {'n': 2, 't': t3, 'tag': b'ad'}])
# HOPING BETWEEN GROUND STATE AND EXCITED STATE
t01 = 2.
sys.set_hopping_manual({(26, 27): t01})
```

```python
plt.lattice(plt_hop=True, c=1, ms=3);
```

```python
sys.get_ham()
print(sys.ham.real.toarray())
sys.get_eig(eigenvec=True)
```
