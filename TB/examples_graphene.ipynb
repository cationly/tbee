# GRAPHENE FLAKES UNDER LINEAR TRIAXIAL STRAIN

## Strained Hexagonal Flakes

```python
lat = grapheneLat()
sys = grapheneSys(lat=lat)
plt = plot(sys=sys)
```

```python
lat.hexagon_zigzag(5)
lat.center()
betas = sys.get_beta_lims()
```

### Increasing negative strain

```python
sys.set_hop_linear_strain(t=1, beta=0)
sys.get_coor_hop()
plt.lattice_hop(ms=5, c=0.4, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=1/4*betas[0])
sys.get_coor_hop()
plt.lattice_hop(ms=5, c=0.4, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=2/4*betas[0]);
sys.get_coor_hop()
plt.lattice_hop(ms=5, c=0.4, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=3/4*betas[0])
sys.get_coor_hop()
plt.lattice_hop(ms=5, c=0.4, plt_hop=True, figsize=(3, 3));
```

### Increasing positive strain

```python
sys.set_hop_linear_strain(t=1, beta=0)
sys.get_coor_hop()
plt.lattice_hop(ms=5, c=0.4, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=1/4*betas[1])
sys.get_coor_hop()
plt.lattice_hop(ms=5, c=0.4, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=2/4*betas[1])
sys.get_coor_hop()
plt.lattice_hop(ms=5, c=0.4, plt_hop=True, figsize=(3, 3));

sys.set_hop_linear_strain(t=1, beta=3/4*betas[1])
sys.get_coor_hop()
plt.lattice_hop(ms=5, c=0.4, plt_hop=True, figsize=(3, 3));
```

## Vacancy defect leading to localized states between Laudau levels 

### site A at the center of mass removed

```python
lat = grapheneLat()
sys = grapheneSys(lat=lat)
plt = plot(sys=sys)

n = 5
lat.triangle_zigzag(n)
lat.center()
lat.remove_sites([lat.sites//3])
#betas = sys.get_beta_lims()
#sys.set_hop_linear_strain(t=1, beta=betas[1])
plt.lattice(ms=5, fs=15, tics=True, plt_index=True);
print(lat.sites//3)
```

```python
lat = grapheneLat()
sys = grapheneSys(lat=lat)
plt = plot(sys=sys)

lat.triangle_zigzag(25)
lat.center()
betas = sys.get_beta_lims()
lat.remove_sites([377])
sys.set_hop_linear_strain(t=1, beta=betas[1])
plt.lattice(ms=5);
```

```python
sys.get_ham()
sys.get_eig(eigenvec=True)
plt.spectrum(ms=8, tag_pola=b'a', fs=15, lims=[-2., 2.]);
```

```python
ll0 = sys.get_intensity_en(lims=[-0.2, 0.2])
ll1 = sys.get_intensity_en(lims=[.8, 1.])
ll2 = sys.get_intensity_en(lims=[1.1, 1.3])
v0 = sys.get_intensity_en(lims=[0.4, 0.6])
v1 = sys.get_intensity_en(lims=[1., 1.1])
v2 = sys.get_intensity_en(lims=[1.35, 1.4])
```

```python
plt.intensity_area(ll0, s=20, plt_hop=True);
plt.intensity_area(ll1, s=20, plt_hop=True);
plt.intensity_area(ll2, s=20, plt_hop=True);
```

```python
plt.intensity_area(v0, s=40, plt_hop=True);
plt.intensity_area(v1, s=40, plt_hop=True);
plt.intensity_area(v2, s=40, plt_hop=True);
```
