# Strain Graphene

Theses examples illustrate the impact of triaxial linear strain on graphene flakes:
- Formations Pseudo Landau levels (pLLs).
- Energies depending on strain strength. 
- Flake geometry influence.
- Vacancy defect induced states between pLLs.
- Negative couplings induced states between pLLs.

## Squared flake

```python
graphene_tb.get_lattice(flake=square, n=20)
plt_graphene = plotGraphene(graphene_tb.coor, graphene_tb.tags)
eig_graphene = eigGraphene(graphene_tb.coor, graphene_tb.coor_rec, graphene_tb.tags)
```

```python
eig_graphene.get_hops(t1, t2, t3, beta=eig_graphene.beta_lims[1], get_hop_rec=True)
```

```python
eig_graphene.get_coor_hop(graphene_tb.nx, graphene_tb.ny,
           graphene_tb.coor_rec, graphene_tb.ind)
fig_lat_hop = plt_graphene.plt_lattice(eig_graphene.coor_hop, ms=8)
```

```python
# butterfly
eig_graphene.get_butterfly(t1, t2, t3, N=10)
fig_butterfly = plt_graphene.plt_butterfly(eig_graphene.butterfly, eig_graphene.betas, 
                                           en_lims=[-2., 2.])
```

```python
import numpy as np
```

```python
import scipy.sparse as sparse
```

```python
n = 10
```

```python
ham = sparse.csr_matrix((np.arange(n), (np.arange(0, n), np.arange(1, n+1))), shape=(n+1, n+1), dtype='f8') \
    + sparse.csr_matrix((np.arange(n), (np.arange(1, n+1), np.arange(0, n))), shape=(n+1, n+1), dtype='f8')
```

```python
print(ham.toarray())
ham 
```

```python
import scipy.sparse.linalg as sla
import scipy.linalg as LA
```

```python
e = sla.eigsh(ham, k=5, sigma=1e-4, return_eigenvectors=False, tol=1e-5, maxiter=5000)
e = np.sort(e)
print(e)
```

```python
print(LA.eigvalsh(ham.toarray()))
```
