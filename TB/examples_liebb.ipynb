# Lieb lattice with 65 sites

```python
n1, n2 = 5, 5
lat = liebLat()
sys = liebSys(lat=lat)
plt = plot(sys=sys)
lat.get_lattice(n1=n1, n2=n2)
lat.remove_dangling()
```

## Uniform hoppings

```python
t = 1.
t_nn = 0.5
sys.set_hopping([{'n': 1, 't': t}, {'n': 2, 't': t_nn}])
plt.lattice(ms=15, plt_hop=True);
```

```python
sys.get_ham()
sys.get_eig(eigenvec=True)
```

```python
plt.spectrum(tag_pola=b'a');
```

```python
zero_mode = sys.get_intensity_pola_max(tag_pola=b'a')
```

```python
plt.intensity_disk(zero_mode);
```

## Alternating Hoppings

```python
t_ab, t_ba = np.round(4/3, 2), np.round(2/3, 2)
t_ac, t_ca = t_ab, t_ba
t_mp, t_pp, t_mm, t_pm = 0.8, 0.4, 0.4, 0.2
sys.set_nearest_neighbor_hopping(t_ab=t_ab, t_ba=t_ba, t_ac=t_ac, t_ca=t_ca)
sys.set_next_nearest_neighbor_hopping(t_pp=t_pp, t_pm=t_pm, t_mp=t_mp ,t_mm=t_mm)
```

```python
plt.lattice(plt_hop=True, ms=10, c=3);
```

```python
sys.get_ham()
sys.get_eig(eigenvec=True)
plt.spectrum(tag_pola=b'a');
```

```python
zero_mode = sys.get_intensity_pola_max(tag_pola=b'a')
```

```python
plt.intensity_disk(zero_mode);
```
