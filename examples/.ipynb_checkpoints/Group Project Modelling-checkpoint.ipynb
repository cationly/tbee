### Graphene with Impurity

```python
lat = grapheneLat()
sys = grapheneSys(lat=lat)
plt = plot(sys=sys)

lat.get_lattice(n1=7, n2=5)
lat.remove_dangling()

lat.remove_sites([30])

sites_no_impurity = lat.sites
coor = np.array([(7, 3, b'c'), (7, 3, b'd')], dtype=[('x', 'f8'), ('y', 'f8'), ('tag', 'S1')])
lat.add_sites(coor=coor)

plt.lattice(axis=True, ms=10, fs=10, plt_index=True);
```

```python
#onsite energy of the two levels of the impurity
sys.set_onsite({b'c': 1, b'd': 5})
```

```python
#setting nearest and next-nearest neighbour hopping in the normal lattice
sys.set_hopping([{'n':1, 't':1., 'n':2, 't':0.5}])
```

```python
#site 67 is the ground state of the impurity, of higher energy than ground state of other lattice atoms
i0 = 67  
t0 = 0.8

#setting hopping from sites 23, 35, 36 to the ground state of the impurity
sys.set_hopping_manual({(23, i0): t0, (35, i0): t0, 
                        (36, i0): t0})

#site 68 is the excited state of the impurity
i1 = 68 
t1 = 2
#setting hopping from sites 23, 35, 36 to the excited state of the impurity
sys.set_hopping_manual({(23, i1): t1, (35, i1): t1, 
                        (36, i1): t1})
```

```python
sys.get_ham()
print(sys.ham.real.toarray())
sys.get_eig(eigenvec=True)
```
