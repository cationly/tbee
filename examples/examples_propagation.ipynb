# Graphene Placket

```python
from math import pi as PI
from math import cos, sin, sqrt
```

```python
unit_cell = [{'tag': b'a', 'r0': (0., 0.)}, 
             {'tag': b'b', 'r0': (0.5, 0.5/sqrt(3))}]
prim_vec = [(1, 0.), 
            (cos(PI/3), sin(PI/3))]
lat = lattice(unit_cell=unit_cell, prim_vec=prim_vec)
n1, n2 = 2, 2
lat.get_lattice(n1=n1, n2=n2)
lat.remove_dangling()

sys = system(lat=lat)
t_1 = 1. 
sys.set_hopping([{'n': 1, 't': t_1}])
sys.get_ham()
```

```python
prop.prop
```

```python
prop = propagation(lat=lat)
psi_init = np.zeros(lat.sites, 'c16')
psi_init[0] = 1.
prop.get_propagation(ham=sys.ham, psi_init=psi_init, 
                     steps=50, dz=0.2, norm=False)
ani = prop.get_animation_nb(s=400., ani_type='norm')
```
